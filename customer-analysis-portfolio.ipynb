```python
# üõ†Ô∏è Technical Stack & Environment Setup
# Portfolio Highlight: Professional data science toolkit implementation

# METHODOLOGY NOTE: This notebook demonstrates production-ready data science practices
# including comprehensive environment setup, systematic feature engineering,
# rigorous statistical validation, and executive-level business intelligence.
# Each section builds upon previous work to create a complete analytics framework.

# Core Data Science Libraries
import pandas as pd              # Data manipulation and analysis
import numpy as np               # Numerical computing

# Machine Learning & Statistics
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from scipy import stats

# Advanced Visualization Suite
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings

# Configuration for professional output
warnings.filterwarnings('ignore')
plt.style.use('default')
sns.set_palette("husl")

print("üöÄ Professional Data Science Environment Initialized")
print("üìä Ready for: ML Clustering | Statistical Validation | Interactive Analytics")
```

```python
# Load and explore the customer dataset
data = pd.read_csv('./marketing_campaign.csv', sep='\t')

print(f"üìä Dataset loaded: {data.shape[0]:,} customers, {data.shape[1]} features")
print(f"üí∞ Revenue analyzed: ${data[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum().sum():,.0f}")

# Display dataset overview
display(data.head())
print("\nüìã Dataset Info:")
print(data.info())
```

## üî¨ Advanced Feature Engineering & Data Preprocessing
# Portfolio Highlight: Scalable ML pipeline design with industry best practices

Strategic data preparation to optimize clustering performance and business interpretability.

```python
# Create a copy for preprocessing
df = data.copy()

# 1. Handle missing values strategically
print("üîç Missing Values Analysis:")
missing_summary = df.isnull().sum()[df.isnull().sum() > 0]
if len(missing_summary) > 0:
    print(missing_summary)
    # Income imputation using education and marital status groups
    df['Income'] = df.groupby(['Education', 'Marital_Status'])['Income'].transform(
        lambda x: x.fillna(x.mean()) if x.notna().sum() > 0 else x.fillna(df['Income'].mean())
    )
else:
    print("‚úÖ No missing values detected")

# 2. Feature engineering for customer analytics
print("\nüõ†Ô∏è Engineering Customer Features:")

# Date processing
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df['Customer_Days'] = (pd.Timestamp.now() - df['Dt_Customer']).dt.days
df['Customer_Years'] = df['Customer_Days'] / 365.25

# Spending aggregations
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['MntTotal'] = df[spending_cols].sum(axis=1)
df['total_spending'] = df['MntTotal']

# Purchase behavior
purchase_cols = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df['total_purchases'] = df[purchase_cols].sum(axis=1)
df['avg_purchase_value'] = df['MntTotal'] / (df['total_purchases'] + 1)  # +1 to avoid division by zero

# Family dynamics
df['family_size'] = df['Kidhome'] + df['Teenhome'] + 1  # +1 for customer
df['spending_per_family_member'] = df['MntTotal'] / df['family_size']

# Customer value metrics
df['customer_lifetime_value'] = df['MntTotal'] * (df['Customer_Years'] + 1)
df['income_per_family_member'] = df['Income'] / df['family_size']

print(f"‚úÖ Created {len(['Customer_Days', 'Customer_Years', 'MntTotal', 'total_spending', 'total_purchases', 'avg_purchase_value', 'family_size', 'spending_per_family_member', 'customer_lifetime_value', 'income_per_family_member'])} new features")

# 3. Data quality checks
print("\nüéØ Data Quality Summary:")
print(f"‚Ä¢ Original dataset: {data.shape[0]:,} customers")
print(f"‚Ä¢ After preprocessing: {df.shape[0]:,} customers")
print(f"‚Ä¢ Features available: {df.shape[1]} total")
print(f"‚Ä¢ Total revenue: ${df['MntTotal'].sum():,.0f}")
print(f"‚Ä¢ Average customer value: ${df['MntTotal'].mean():.0f}")
```

## ü§ñ Machine Learning Excellence & Model Optimization
# Portfolio Highlight: Production-ready clustering with rigorous validation framework

Applying K-Means clustering with statistical validation to identify optimal customer segments.

```python
# Prepare features for clustering
feature_cols = ['Income', 'MntTotal', 'total_purchases', 'avg_purchase_value', 'family_size', 
                'Customer_Years', 'Age', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# Create feature matrix
X = df[feature_cols].copy()

# Handle any remaining missing values
X = X.fillna(X.mean())

# Standardize features for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"üéØ Clustering Features Prepared:")
print(f"‚Ä¢ Features selected: {len(feature_cols)}")
print(f"‚Ä¢ Customers for analysis: {X_scaled.shape[0]:,}")
print(f"‚Ä¢ Feature matrix shape: {X_scaled.shape}")

# Test multiple cluster numbers to find optimal k
k_range = range(2, 8)
silhouette_scores = []
davies_bouldin_scores = []
inertias = []

print("\nüîç Evaluating Optimal Number of Clusters:")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    sil_score = silhouette_score(X_scaled, labels)
    db_score = davies_bouldin_score(X_scaled, labels)
    
    silhouette_scores.append(sil_score)
    davies_bouldin_scores.append(db_score)
    inertias.append(kmeans.inertia_)
    
    print(f"k={k}: Silhouette={sil_score:.4f}, Davies-Bouldin={db_score:.2f}")

# Find optimal k
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"\nüèÜ Optimal k selected: {optimal_k} (Silhouette Score: {max(silhouette_scores):.4f})")
```

```python
# Visualize cluster validation metrics
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Cluster Validation Analysis', fontsize=16, fontweight='bold')

# Elbow curve
ax1.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia (Within-cluster sum of squares)')
ax1.set_title('Elbow Method')
ax1.grid(True, alpha=0.3)

# Silhouette scores
colors = ['red' if k == optimal_k else 'blue' for k in k_range]
ax2.bar(k_range, silhouette_scores, color=colors, alpha=0.7)
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title(f'Silhouette Analysis (Optimal k={optimal_k})')
ax2.grid(True, alpha=0.3)

# Davies-Bouldin scores (lower is better)
ax3.plot(k_range, davies_bouldin_scores, 'go-', linewidth=2, markersize=8)
ax3.set_xlabel('Number of Clusters (k)')
ax3.set_ylabel('Davies-Bouldin Score (lower is better)')
ax3.set_title('Davies-Bouldin Analysis')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üìä Statistical Validation Results:")
print(f"‚Ä¢ Optimal clusters: {optimal_k}")
print(f"‚Ä¢ Silhouette score: {max(silhouette_scores):.4f} (higher is better)")
print(f"‚Ä¢ Davies-Bouldin score: {davies_bouldin_scores[optimal_k-2]:.2f} (lower is better)")
```

```python
# Apply optimal K-Means clustering
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = final_kmeans.fit_predict(X_scaled)

# Add cluster labels to dataframe
df['Cluster'] = cluster_labels

# Calculate final validation metrics
final_silhouette = silhouette_score(X_scaled, cluster_labels)
final_davies_bouldin = davies_bouldin_score(X_scaled, cluster_labels)
final_calinski_harabasz = calinski_harabasz_score(X_scaled, cluster_labels)

print(f"üéØ Final K-Means Model Performance:")
print(f"‚Ä¢ Clusters: {optimal_k}")
print(f"‚Ä¢ Customers clustered: {len(cluster_labels):,}")
print(f"‚Ä¢ Silhouette Score: {final_silhouette:.4f}")
print(f"‚Ä¢ Davies-Bouldin Score: {final_davies_bouldin:.2f}")
print(f"‚Ä¢ Calinski-Harabasz Score: {final_calinski_harabasz:.2f}")

# Cluster size distribution
cluster_sizes = df['Cluster'].value_counts().sort_index()
print(f"\nüë• Cluster Distribution:")
for cluster, size in cluster_sizes.items():
    percentage = (size / len(df)) * 100
    print(f"‚Ä¢ Cluster {cluster}: {size:,} customers ({percentage:.1f}%)")
```
